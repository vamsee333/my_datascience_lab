```python
def calc_tip(s):
    if (s.tip/s.total_bill)*100 > 20:
        print(f'{s.total_bill} and {s.tip} -High')
    elif (s.tip/s.total_bill)*100 > 10:
        print(f'{s.total_bill} and {s.tip} -avg')
    else:
        print(f'{s.total_bill} and {s.tip}-low')
```
